```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("banking_data.csv")
```

```python
df.head(5)
```

### Plot 1: Wait time vs Order Price

```python
plt.figure(figsize=(15, 5))
plt.subplot(131)
plt.hexbin(df['Wait_Time_Minutes'], df['Order_Price'], gridsize=10, cmap='YlOrRd')
plt.colorbar(label='Count')
plt.xlabel('Wait Time (Min)')
plt.ylabel('Order Price ($)')

plt.title('Wait time vs Order Price')
```

### Plot 2: Visit Count vs Order Price

```python
plt.figure(figsize=(15, 5))
plt.subplot(132)
plt.hexbin(df['Visit_Count'], df['Order_Price'], gridsize=10, cmap='YlOrRd')
plt.colorbar(label='Count')
plt.xlabel('Visit Count')
plt.ylabel('Order Price ($)')

plt.title('Visit Count vs Order Price')
```

### Plot 3: Visit Count vs Wait Time

```python
plt.figure(figsize=(15, 5))
plt.subplot(133)
plt.hexbin(df['Visit_Count'], df['Wait_Time_Minutes'], gridsize=10, cmap='YlOrRd')
plt.colorbar(label='Count')
plt.xlabel('Visit Count')
plt.ylabel('Wait_Time_Minutes')

plt.title('Visit Count vs Wait_Time_Minutes')
```
