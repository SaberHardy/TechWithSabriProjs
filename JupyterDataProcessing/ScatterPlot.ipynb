```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
x = np.random.rand(100)
y = np.random.rand(100)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='blue', alpha=0.5)
plt.title("Simple Scatter Plot of Random Data")
plt.xlabel("Years")
plt.ylabel("Sales")
plt.show()
```

### Scatter plot using Reak Data (Tip Dataset)

```python
tips_data = sns.load_dataset('tips')
tips_data.head(3)
```

```python
"""
total_bill: the total bill ampiunt in dollars
tip: the tip ampunt in dollars
sex: the gender og the person paying the bill (Male | Female)
smoker: whether the person paying the bill is a smoker or not
day: the day of the week
time: the time of the day(lunch or dinner)
size: the nbr of the people in the party
"""
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(x='total_bill', y='tip', data=tips_data, color='blue')
plt.title("Scatter Plot of Total bills vs Tip")
plt.xlabel("Total Bill ($)")
plt.ylabel("Tip ($)")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='total_bill', y='tip', size='size', hue='day', data=tips_data, color='blue')
plt.title("Scatter Plot of Total bills vs Tip with Size and Color Mapping")
plt.xlabel("Total Bill ($)")
plt.ylabel("Tip ($)")
plt.show()
```

### Adding a Regression Line

```python
plt.figure(figsize=(10, 6))
sns.regplot(x='total_bill', y='tip', data=tips_data,scatter_kws={'color':'blue'}, line_kws={'color': 'red'})
plt.title("Scatter Plot of Total bills vs Tip Regression line")
plt.xlabel("Total Bill ($)")
plt.ylabel("Tip ($)")
plt.show()
```
