```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
data = np.random.randn(100)
```

```python
plt.figure(figsize=(10, 6))
plt.boxplot(data, patch_artist=True, boxprops=dict(facecolor='lightblue'))
plt.title("Simple Box plot with random data")
plt.ylabel("Values")
plt.show()
```

### Box Plot using Tips dataset

```python
tips = sns.load_dataset('tips')
tips.head(2)
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='day', y='total_bill', data=tips, palette='Set3', hue='day')
plt.title("Box Plot of total bills amount by day")
plt.xlabel("Day")
plt.ylabel("Total Bill $")
plt.show()
```

### Box plot with multiple Categories (Hue Paramete)

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='day', y='total_bill',hue='sex', data=tips, palette='coolwarm')
plt.title("Box Plot of total bills amount by day with Gender Differentiation")
plt.xlabel("Day")
plt.ylabel("Total Bill $")
plt.legend(title='Gender')
plt.show()
```

### Horizontal Box Plot With Outliers Highlighted

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='tip', data=tips, orient='h', flierprops=dict(marker='o', color='red', markersize=8))
plt.title("Horizontal Box Plot of Tips with Highlited Outliers")
plt.xlabel("Tip Amount ($)")
plt.show()
```
