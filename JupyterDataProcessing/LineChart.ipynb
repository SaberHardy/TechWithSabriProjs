```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = {
    'Month': ['Jan','Feb','Mar','Apr', 'May'],
    'Sales': [150, 200, 300, 200, 400]
}
df = pd.DataFrame(data)
df
```

```python
plt.plot(df['Month'], df['Sales'], marker='*', linestyle='-', color='r')
plt.title("Monthly Sales")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```
