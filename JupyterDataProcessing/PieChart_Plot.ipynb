```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
```

```python
df = pd.read_csv("banking_data.csv")
df.head(3)
```

```python
df['Item_Ordered'].unique()
```

```python
plt.subplot(221)
rating_count = df['Rating'].value_counts()
# rating_count
```

```python
plt.pie(rating_count, labels=rating_count.index, autopct='%1.1f%%',
       colors=sns.color_palette('pastel'),
        explode=[0.05] * len(rating_count))
plt.title("Distribution of Customer Ratings")
plt.show()
```

### Service Quality Pie Chart

```python
service_quality = df['Service_Quality'].value_counts()
plt.pie(service_quality, labels=service_quality.index, autopct='%1.1f%%',
       colors=sns.color_palette('Set3'),
        explode=[0.05] * len(service_quality))
plt.title("Distribution of Service Quality")
plt.show()
```

```python
item_ordered = df['Item_Ordered'].value_counts()
plt.pie(item_ordered, labels=item_ordered.index, autopct='%1.1f%%',
       colors=sns.color_palette('Set3'),
        explode=[0.05] * len(item_ordered))
plt.title("Distribution of Item Ordered")
plt.show()
```

### Custom Price Range Pie Chart

```python
price_range = df['Order_Price'].value_counts()
plt.pie(price_range, labels=price_range.index, autopct='%1.1f%%',
       colors=sns.color_palette('Set3'),
        explode=[0.05] * len(price_range))
plt.title("Distribution of Price Range")
plt.show()
```
