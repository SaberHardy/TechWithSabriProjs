```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
data = np.random.randn(100)
data
```

```python
plt.figure(figsize=(10, 6))
plt.hist(data, bins=30, color='skyblue', edgecolor='red')
plt.title("Simple Histogram of Random Data")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

### Histogram of Real Data (Tips Data)

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
plt.figure(figsize=(10, 6))
plt.title("Histogram of total Bill Amounts")
sns.histplot(tips['total_bill'], bins=20, kde=True, color='green')
plt.xlabel('Total Bill $')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.histplot(tips['total_bill'], 
             bins=20,
             alpha=0.5,
             color='orange',
             edgecolor='black')

sns.histplot(tips['tip'], 
             bins=20,
             alpha=0.5,
             color='blue',
             edgecolor='black')
plt.title("Comparision of total bill and tip amounts")
plt.xlabel('Total Bill $')
plt.ylabel('Frequency')
plt.show()
```

```python
data_exp = np.random.exponential(scale=2.0, size=1000)
data_exp
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data_exp, 
             bins=20,
             color='green',
             edgecolor='yellow')
plt.yscale('log')
plt.title("Histogram with log scale .... ")
plt.xlabel('Value')
plt.ylabel('Log Frequency')
plt.show()
```
