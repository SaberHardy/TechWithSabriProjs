```python
import pandas as pd
import numpy as np
```

```python
data = {
    'Car': ['Toyota','Ford',None,'Honda','Tesla','Audi'],
    'Sales': [np.nan, 45, 546, None, 200, 788],
    'Year': [2000, 2020, None, 2023, 1999, 2040],
    'Quarter': ['Q1',None,'Q2','Q1','Q3','Q1']
}
data
```

```python
df = pd.DataFrame(data)
df
```

```python
df_dropna = df.dropna()
df_dropna
```

```python
# Drop columns with any missing values
df_dropna_cols = df.dropna()
df_dropna_cols
```

```python
# Fill Numiric values with the column's mean
df_filled_by_mean = df.copy()
df_filled_by_mean['Sales'].fillna(df['Sales'].mean(), inplace=True)
df_filled_by_mean
```

```python
# Fill missing categorical values with specific value ('Unknown' for 'Car' column)
df_filled_by_unknown = df.copy()
df_filled_by_unknown
```

```python
df_filled_by_unknown['Car'].fillna('Unknown', inplace=True)
df_filled_by_unknown
```

```python
# Forward-fill missing values
df_forward_fill = df.copy()
df_forward_fill
```

```python
df_forward_fill['Car'].fillna('Unknown', inplace=True)
df_forward_fill
```

```python
df_mixed_fill = df.copy()
df_mixed_fill
```

```python
df_mixed_fill['Car'].fillna('Unknown', inplace=True)
df_mixed_fill['Sales'].fillna(df['Sales'].median(), inplace=True)
df_mixed_fill['Quarter'].fillna('Q4', inplace=True)
df_mixed_fill
```
