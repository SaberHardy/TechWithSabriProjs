```python
# Conclusion: 
# 0: drop rows
# 1: drop Columns
```

```python
# Drop Duplication
df = pd.DataFrame({
    'name': ['name1', 'name2', 'name3', 'name2', 'name3'],
    'age': [12,34,54,21,34],
    'country': ['alg', 'uk', 'aus', 'ind', 'alg']
})
df
```

```python
df['name'].duplicated()
```

```python
dropped = df.drop_duplicates(subset=['country'])
dropped
```

```python
# Drop null values
df = pd.DataFrame({
    'name': ['name1', np.nan, 'name3', 'name2', 'name3'],
    'age': [12, np.nan,54,21,34],
    'country': ['alg', 'uk', 'aus', np.nan, 'alg']
})
df
```

```python
dropped = df.dropna()
dropped
```

```python
# Conlusion
# 1. labels: name of the row/column to drop
# 2. axis: 0: For rows, 1: For Columns
# 3. inplace: Activate dropping the column for 
```
